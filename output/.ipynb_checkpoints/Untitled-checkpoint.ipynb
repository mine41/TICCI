```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scanpy as sc
import anndata as ad
from sklearn.preprocessing import  PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
test = pd.read_csv('./AT2/100全部测试.csv',index_col = 0)
test
```

```python
test['PAS'] = test['PAS'] / 0.676831683
test['PAS']
```

```python
test['PAS'].values.tolist()
```

```python
test.index.values.tolist()
```

```python
## AT2
x_values = test.index.values.tolist()
y_values = test['PAS'].values.tolist()

max_y = max(y_values)
max_x = x_values[y_values.index(max_y)]
zero_y = y_values[0]

plt.figure(figsize=(12,4))
plt.plot(x_values,y_values)

plt.vlines(max_x,0,max_y,linestyles='dashed',colors='red')
plt.hlines(max_y,-3,max_x,linestyles='dashed',colors='red')
plt.text(max_x, 0.782,max_x,ha='center')
plt.text(-1.1, max_y+0.005,round(max(y_values),4),va='center')

plt.vlines(0,0,zero_y,linestyles='dashed',colors='red')
plt.hlines(zero_y,-3,0,linestyles='dashed',colors='red')
plt.text(-1.1, zero_y,round(zero_y,4),va='center')

plt.ylim(0.79,0.91)
plt.xlim(-0.5,10.2)

plt.xlabel("CCI_weight")
plt.ylabel("PAS")
plt.title("B.AT2",fontsize=25)

plt.savefig("AT2_CCI_weight.svg")
plt.show()
```

```python
# #AT2
data = [0.9005, 0.7934, 0.9087, 0.9327, 0.6512]
x_axis_labels = ['TICCI(ours)','PAGA','TSCAN','Monocle2','Monocle3']
fig, ax = plt.subplots(figsize=(8,4))
ax.set_ylim([0, 1.0])

plt.bar(x_axis_labels, data)

for i in range(len(data)):
    plt.text(i-0.25, data[i]+0.01, str(data[i]))

plt.title("B.AT2",fontsize=25)
plt.ylabel('PAS')
plt.savefig('AT2_other.svg', format='svg')
plt.show()
```
